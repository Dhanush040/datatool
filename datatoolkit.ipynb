Understanding NumPy Arrays and Copy Operations
np.array, np.asarray, and np.asanyarray
np.array:

Creates a new NumPy array from a sequence or other array-like object.
Copies the data by default.
Example: arr = np.array([1, 2, 3])

np.asarray:

Converts an input to a NumPy array.
If the input is already a NumPy array, it returns the same array without copying.
If the input is not a NumPy array, it creates a new NumPy array.
Example: arr = np.asarray([1, 2, 3])
np.asanyarray:

Similar to np.asarray but allows for subclasses of NumPy arrays.
If the input is a subclass, it returns the subclass without copying.Otherwise, it behaves like np.asarray.
Example: arr = np.asanyarray([1, 2, 3])
Deep Copy vs. Shallow Copy

Shallow Copy:

Creates a new array object but shares the same data with the original array.
Changes to the original array will affect the copy.
Use np.copy to create a shallow copy.
Example: arr_copy = arr.copy()

Deep Copy:
Creates a new array object with a completely independent copy of the data.
Changes to the original array will not affect the copy.
Use np.deepcopy to create a deep copy.
Example: import copy; arr_copy = copy.deepcopy(arr)


```python
import numpy as np
import copy

arr = np.array([1, 2, 3])

# Shallow copy
arr_copy1 = arr.copy()
arr[0] = 10  # Changes both arr and arr_copy1

# Deep copy
arr_copy2 = copy.deepcopy(arr)
arr[0] = 20  # Changes only arr, not arr_copy2
```

```python
#15 Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
#a) Plot a histogram of the data with 30 bins.

#b) Overlay a line Plot representing the normal distribution's probability density function (PDF).

#c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

#d) Set the title of the plot as 'Histogram with PDF Overlay
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create 1000 samples from a normal distribution
data = np.random.randn(1000)

# Calculate mean and standard deviation of the data
mean, std = np.mean(data), np.std(data)

# Create a histogram with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, label='Histogram')

# Create x-axis values for the PDF
x = np.linspace(min(data), max(data), 100)

# Plot the PDF as a line
plt.plot(x, norm.pdf(x, mean, std), 'r-', lw=2, label='PDF')

# Set labels and title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

# Add legend
plt.legend()

plt.show()

```

```python
#17 Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate two random arrays
np.random.seed(0)  # For reproducibility
x = np.random.randn(100)
y = np.random.randn(100)

# Define a function to categorize points based on quadrants
def categorize_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Quadrant I'
    elif x < 0 and y > 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

# Apply the function to the arrays
categories = [categorize_quadrant(xi, yi) for xi, yi in zip(x, y)]

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=x, y=y, hue=categories, palette='tab10', legend='full')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Show plot with legend
plt.legend(title='Quadrants')
plt.grid(True)
plt.show()
```

```python
#18 With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function
import numpy as np
from bokeh.plotting import figure, show

# Create data for the sine wave
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Create the plot
p = figure(title="Sine Wave Function", x_axis_label="x", y_axis_label="sin(x)")

# Add the line plot
p.line(x, y, line_width=2)

# Add grid lines
p.grid.grid_line_alpha = 0.3

# Show the plot
show(p)
```

```python
#19 Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on theirvalues, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool, CategoricalColorMapper
from bokeh.palettes import Category20

# Generate random categorical data
categories = ["A", "B", "C", "D", "E"]
values = np.random.randint(1, 20, size=len(categories))

# Create color map
color_mapper = CategoricalColorMapper(palette=Category20[len(categories)], factors=categories)

# Create data source
source = ColumnDataSource(data=dict(
    categories=categories,
    values=values,
    color=categories
))

# Create the plot
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label="Category", y_axis_label="Value")

# Add the bar plot
p.vbar(x='categories', top='values', width=0.5, color='color', source=source)

# Add hover tool
hover = HoverTool(tooltips=[("Category", "@categories"), ("Value", "@values")])
p.add_tools(hover)

# Show the plot
show(p)
```
